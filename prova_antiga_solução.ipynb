```python
#questão 1

import numpy as np
import matplotlib.pyplot as plt

A = np.array([[-9.0,32,8],
            [9,12,-4],
             [-12,4,28]])

b = np.array([[227.0],[127],[56]])

solucao = np.linalg.solve(A,b)

print('solução  \n',solucao)

print ('Matriz A \n',A,'\n Matriz b \n',b)

Ab = np.concatenate((A,b),1)

n,_ = np.shape(Ab)

print ('Matriz Ab \n',Ab)

#pivotando


L0old = np.copy(Ab[0,:])
Ab[0,:] = Ab[2,:]
Ab[2,:] = L0old

print ('Matriz Ab \n',Ab)

#segunda linha, primeira coluna

pivo = Ab[0,0]
fator21 = Ab[1,0]/pivo
print ('pivo \n',pivo,'\nfator 21 \n',fator21)
Ab[1,:] = Ab[1,:] - fator21*Ab[0,:]
print(Ab)

#terceira linha, primeira coluna
fator31 = Ab[2,0]/pivo
print ('pivo \n',pivo,'\nfator 31 \n',fator31)
Ab[2,:] = Ab[2,:] - fator31*Ab[0,:]

print(Ab)

L2old = np.copy(Ab[1,:])
Ab[1,:] = Ab[2,:]
Ab[2,:] = L2old

print('2 pivotação \n',Ab)

#terceira linha,segunda coluna

pivo = Ab[1,1]
fator32 = Ab[2,1]/pivo
print ('pivo \n',pivo,'\nfator 32 \n',fator32)
Ab[2,:] = Ab[2,:] - fator32*Ab[1,:]


print(Ab)

#resolvendo o sistema
# separando as matrizes
U = Ab[:,:n]
bi = Ab[:,n:]

print('matriz U \n',U)
print('matriz b1 \n',bi)

sol = np.linalg.solve(U,bi)

print(sol)
```
